```python
import numpy as np
import pandas as pd
import json

class Operator:
    def __init__(self, operator_hash, t, operator):
        self.operator_hash = operator_hash
        self.t = t
        self.operator = operator
        self.output_operator = None
        self.left_input_operator = None
        self.right_input_operator = None
        
def read_operator(operator_type):
    return [Operator(r.OPERATOR_HASH, operator_type, r) for r in pd.read_csv(benchmark + '/' + operator_type + '.csv', sep='|').itertuples()]

def read_operators():
    operator_types = ['get_tables', 'aggregates', 'joins', 'projections', 'table_scans', 'validates']
    operators = []
    for op in operator_types:
        operators += read_operator(op)

def build_tree(operators):
    hashes_to_operators = {op.operator_hash: op for op in operators}
    for op in operators:
        h = op.operator.LEFT_INPUT_OPERATOR_HASH
        if h is not None:
            o = hashes_to_operators.get(h)
            if o is not None:
                o.output_operator = op
                op.left_input_operator = o
        h = op.operator.RIGHT_INPUT_OPERATOR_HASH
        if h is not None:
            o = hashes_to_operators.get(h)
            if o is not None:
                o.output_operator = op
                op.right_input_operator = o
    
def keeps_sort_order(op):
    if op == 'get_tables':
        return True
    elif op == 'aggregates':
        return False
    elif op == 'joins':
        return False
    elif op == 'projections':
        return False
    elif op == 'table_scans':
        return False
    elif op == 'validates':
        return False
    
def compute_join_sorting(operators):
    joins = [op for op in operators if op.t == 'joins']
    for join in joins:
        break
        #if keeps_sort_order()


def run():
    operators = read_operators()
    query_hashes = np.unique(np.array([op.operator.QUERY_HASH for op in operators]))
    print(query_hashes)
    build_tree(operators)
run()
```
